```python
#Data Loading and Preprocessing
import pandas as pd
import numpy as np
import ast

df = pd.read_csv("imdb_data (1).csv")
df.columns = df.columns.str.strip().str.lower()

df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')

df['profit'] = df['revenue'] - df['budget']
df['roi'] = df['profit'] / df['budget']

df = df[df['budget'] > 0]

def get_director(crew_data):
    try:
        crew_list = ast.literal_eval(crew_data)
        for person in crew_list:
            if isinstance(person, dict) and person.get('job') == 'Director':
                return person.get('name')
    except:
        return None

df['director'] = df['crew'].apply(get_director)

df.head()
```

```python
#Q1. Highest Profit movie
top_movie = df.loc[df['profit'].idxmax()]

top_movie[['title','profit','production_companies','director','cast']]
```

```python
#Q2. Language with Highest Average ROI
roi_language = df.groupby('original_language')['roi'].mean().sort_values(ascending=False)

roi_language.head(1)
```

```python
#Q3. Unique genres
unique_genres = set()

for entry in df['genres'].dropna():
    for g in str(entry).split(','):
        unique_genres.add(g.strip())

unique_genres
```

```python
#Q4. Top Producers by ROI
producer_roi = (
    df.groupby('production_companies')['roi']
    .mean()
    .sort_values(ascending=False)
    .head(3)
)

producer_roi
```

```python
#Q5. Actor with Most Movies
actor_count = {}

for cast_list in df['cast'].dropna():
    for actor in str(cast_list).split(','):
        actor = actor.strip()
        actor_count[actor] = actor_count.get(actor, 0) + 1

top_actor = max(actor_count, key=actor_count.get)

top_actor
```

```python
#Q6. Actor Analysis
actor_data = df[df['cast'].str.contains(top_actor, na=False)]

actor_data[['title','profit','genres']].head()
```

```python
#Q6. Directors and their Preferred Actors
director_actor_map = {}

for _, row in df.iterrows():
    director = row['director']
    if pd.isna(director):
        continue

    cast_members = str(row['cast']).split(',')

    if director not in director_actor_map:
        director_actor_map[director] = {}

    for actor in cast_members:
        actor = actor.strip()
        director_actor_map[director][actor] = (
            director_actor_map[director].get(actor, 0) + 1
        )

top_directors = df['director'].value_counts().head(3).index

for d in top_directors:
    print("\nDirector:", d)
    print(sorted(director_actor_map[d].items(), key=lambda x: x[1], reverse=True)[:3])
```
